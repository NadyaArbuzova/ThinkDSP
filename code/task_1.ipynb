```python
from thinkdsp import read_wave

wave = read_wave('105977__wcfl10__favorite-station.wav')
wave.normalize()
wave.make_audio()
```

```python
wave.plot()
```

```python
segment = wave.segment(start=0.5, duration=0.5)
segment.make_audio()
```

```python
segment.plot()
```

```python
segment.segment(start=0.7, duration=0.01).plot()
```

```python
spectrum = segment.make_spectrum()
spectrum.plot(high=8000)
```

```python
from thinkdsp import SinSignal

signal = (SinSignal(freq=300, amp=1.0) +
          SinSignal(freq=500, amp=1.0) +
          SinSignal(freq=600, amp=0.8))
signal.plot()
```

```python
wave2 = signal.make_wave(duration=1)
wave2.apodize()
```

```python
wave2.make_audio()
```

```python
spectrum = wave2.make_spectrum()
spectrum.plot(high=1000)
```

```python
signal += SinSignal(freq=250)
signal.make_wave().make_audio()
```

```python
wave3 = read_wave('105977__wcfl10__favorite-station.wav')
wave3.normalize()
wave3.make_audio()
```

```python
def stretch(wave, factor):
    wave.ts *= factor
    wave.framerate /= factor
```

```python
stretch(wave3, 2.0)
wave3.make_audio()
```
